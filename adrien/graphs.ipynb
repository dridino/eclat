```python
import matplotlib.pyplot as plt
```

```python
d = {}
with open("benchmark.md") as f:
    for i, e in enumerate(f.readlines()):
        if i > 3:
            vals = e.split("|")[1:-1]
            d[vals[0].strip()] = [int(v) for v in vals[1:]]
```

```python
ticks = [1, 2, 4, 8]
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_figwidth(12)
for k, v in list(d.items())[:-2]:
    ax1.plot(ticks, v, label=k)
ax1.set_xlabel("Parallelisation degree")
ax1.set_ylabel("Execution time")
ax1.set_xscale("log", base=2)
ax1.legend()

for k, v in list(d.items())[-2:]:
    ax2.plot([2, 3, 4], v[1:], label=k)
ax2.set_xlabel("Number of piped functions")
ax2.set_ylabel("Execution time")
ax2.set_xticks([2, 3, 4])
ax2.set_yticks([0, 1000, 2000, 3000, 4000])
ax2.legend()
plt.show()
```
